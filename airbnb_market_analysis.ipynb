# performing data cleaning

```python
#checking for missing values.
```

```python
#Cleaning the dataset by removing missing values in one of the key columns "name"
cleaned_data1=df.dropna(subset="name")
cleaned_data1
```

```python
cleaned_data2=cleaned_data1.dropna(subset="reviews_per_month")
cleaned_data2
```

```python
#checking for missing values by host_name
```
